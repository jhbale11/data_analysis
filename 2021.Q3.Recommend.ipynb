# Top 150 Fund's Recommendation 2021 Q3

2021년 Q3에 발행된 상위 150개의 Hedge Fund의 Recommendation을 수합하여 정리하고자 합니다.

사용한 데이터는 https://www.insidermonkey.com/blog/hedge-fund-investor-letters-q3-2021-986500/ 에서 크롤링한 텍스트 데이터입니다.

Hedge Fund들은 2021년 Q3에 어느 주식을 추천하고 있는지 현황 분포를 파악하고자 합니다.

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('fund.txt', sep="(", header=None, names=["fund", "list"])
df
```

```python
df['list'] = df.list.str.split(')').str[0]
df
```

```python
df['list1'] = df.list.str.split(',').str[0]
df['list2'] = df.list.str.split(',').str[1]
df['list3'] = df.list.str.split(',').str[2]
df['list4'] = df.list.str.split(',').str[3]
df['list5'] = df.list.str.split(',').str[4]
df['list6'] = df.list.str.split(',').str[5]
df
```

```python
df = df.drop(['list'], axis=1)
df
```

```python
list1 = df.list1.value_counts().rename_axis('stock').reset_index(name='counts')
list2 = df.list2.value_counts().rename_axis('stock').reset_index(name='counts')
list3 = df.list3.value_counts().rename_axis('stock').reset_index(name='counts')
list4 = df.list4.value_counts().rename_axis('stock').reset_index(name='counts')
list5 = df.list5.value_counts().rename_axis('stock').reset_index(name='counts')
list6 = df.list6.value_counts().rename_axis('stock').reset_index(name='counts')
```

```python
list1
```

```python
total = pd.concat([list1,list2,list3,list4,list5,list6], join='outer')
total
```

```python
total = total.groupby('stock').sum()
total
```
